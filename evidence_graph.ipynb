<font size="6">Step1: seperate the evidences into nodes and get the same line relationship edge</font>

```python
def transfer_graph_step1_sameline(input_file, step1_output_file):
    with open(input_file, 'r', newline='', encoding='utf-8') as result_file:
        
        result_reader = csv.DictReader(result_file)
        
        # 初始化step1的grapgh的csv文件的字段和数据结构
        fieldnames = ['node_evidence', 'evidence_type', 'edge_relationship_sameline','file_name', 'storage_path']
               
        with open(step1_output_file, 'w', newline='', encoding='utf-8') as graph_file:
            # 使用DictWriter写入新csv文件
            result_writer = csv.DictWriter(graph_file, fieldnames=fieldnames)
            result_writer.writeheader()
            
            # 遍历input_file(langchain生成的csv文件)中的每一行,并分析该行中的所有evidence，并生成点
            for row in result_reader:
                evidence_name_lines = row['evidence_name'].split('\n')
                evidence_address_lines = row['evidence_address'].split('\n')
                evidence_phone_number_lines = row['evidence_phone_number'].split('\n')
                evidence_email_lines= row['evidence_email'].split('\n')

                # 处理edge_relationship_sameline的值
                
                #traverse name evidence
                for name_line in evidence_name_lines:
                    sameline_values = []  # 用于存储同一行的数据
                    if(name_line=="None"):
                        continue
                    sameline_values = sameline_value.edge_sameline_combine(name_line,evidence_name_lines,evidence_address_lines,
                                                                           evidence_phone_number_lines, evidence_email_lines)



                    # for other_name in evidence_name_lines:  ##遍历其它的name evidence，并配对成edge
                    #     if(other_name=="None" or other_name==name_line):
                    #         continue
                    #     sameline_values.append((name_line, other_name))                       
                   
                    # for address_line in evidence_address_lines:   ##遍历address evidence，并配对成edge
                    #     if(address_line=="None"):
                    #         continue                       
                    #     sameline_values.append((name_line, address_line))
                    
                    # for phone_number_line in evidence_phone_number_lines:   ##遍历address evidence，并配对成edge
                    #     if(phone_number_line=="None"):
                    #         continue                       
                    #     sameline_values.append((name_line, phone_number_line))
                    
                    # for email_line in evidence_email_lines:   ##遍历address evidence，并配对成edge
                    #     if(email_line=="None"):
                    #         continue                       
                    #     sameline_values.append((name_line, email_line))                    
                    
                    
                    result_writer.writerow({
                        'node_evidence': name_line,
                        'evidence_type': "name",
                        'edge_relationship_sameline': sameline_values,
                        'file_name': row['file_name'],
                        'storage_path': row['storage_path']
                    })                  
                
                #traverse address evidence
                for address_line in evidence_address_lines:
                    sameline_values = []  # 用于存储同一行的数据
                    if(address_line=="None"):
                        continue
                    sameline_values = sameline_value.edge_sameline_combine(address_line,evidence_address_lines,evidence_name_lines,
                                                                           evidence_phone_number_lines, evidence_email_lines)
                    
                    result_writer.writerow({
                        'node_evidence': address_line,
                        'evidence_type': "address",
                        'edge_relationship_sameline': sameline_values,
                        'file_name': row['file_name'],
                        'storage_path': row['storage_path']
                    })  
                
                #traverse phone number evidence 
                for phone_number_line in evidence_phone_number_lines:
                    sameline_values = []  # 用于存储同一行的数据
                    if(phone_number_line=="None"):
                        continue
                    sameline_values = sameline_value.edge_sameline_combine(phone_number_line,evidence_phone_number_lines,
                                                                           evidence_name_lines,
                                                                           evidence_address_lines,
                                                                           evidence_email_lines)
                    
                    result_writer.writerow({
                        'node_evidence': phone_number_line,
                        'evidence_type': "phone number",
                        'edge_relationship_sameline': sameline_values,
                        'file_name': row['file_name'],
                        'storage_path': row['storage_path']
                    })
                
                #traverse email evidence
                
                for email_line in evidence_email_lines:
                    sameline_values = []  # 用于存储同一行的数据
                    if(email_line=="None"):
                        continue
                    sameline_values = sameline_value.edge_sameline_combine(email_line,evidence_email_lines,
                                                                           evidence_name_lines,
                                                                           evidence_address_lines,
                                                                           evidence_phone_number_lines
                                                                           )
                    
                    result_writer.writerow({
                        'node_evidence': email_line,
                        'evidence_type': "email",
                        'edge_relationship_sameline': sameline_values,
                        'file_name': row['file_name'],
                        'storage_path': row['storage_path']
                    })

                    ##改动到这里
 
```

<font size="6">Step2: Get the same table (same file() relationship edge and the same folder relationship edge</font>

```python

def transfer_graph_step2_sametable_samedatabase(input_file, step2_output_file):
    with open(input_file, 'r', newline='', encoding='utf-8') as result_file:
        reader = csv.DictReader(result_file)
        # 初始化step2的grapgh的csv文件的字段和数据结构
        fieldnames = ['node_evidence', 'evidence_type','edge_relationship_sameline', 'edge_relationship_sametable', 'edge_relationship_samedatabase']
        with open(step2_output_file, 'w', newline='', encoding='utf-8') as graph_file:
            # 使用DictWriter写入新csv文件
            result_writer = csv.DictWriter(graph_file, fieldnames=fieldnames)
            result_writer.writeheader()
            
        # 遍历input_file中的每一行,这里的输入是step1的output的csv文件    
            for current_index, current_row in enumerate(reader):
                sametable_values = []  # 用于存储同一文件的数据
                samedatabase_values = [] # 用于存储同一文件夹的数据
                current_file_name = current_row['file_name']
                current_evidence_name = current_row['node_evidence']
                current_storage_path_name = current_row['storage_path'] 
                current_folder_name = os.path.dirname(current_storage_path_name)
                # 从头遍历文件寻找file name一样的行，以及寻找folder一样的行(从storage_path中找)
                
                with open(input_file, 'r', newline='', encoding='utf-8') as second_loop:   ##注意：嵌套遍历循环要再次打开csv文件并设置新的读取器对象reader2
                    reader2 = csv.DictReader(second_loop)
                    for next_index, next_row in enumerate(reader2):
                        next_file_name = next_row['file_name']
                        next_evidence_name = next_row['node_evidence']
                        next_storage_path_name = next_row['storage_path'] 
                        next_folder_name = os.path.dirname(next_storage_path_name)
                        
                        #如果遍历到的是和当前的同一行，则跳过
                        if(next_index==current_index):
                            continue                
                        # 如果遍历到的行的file_name与当前行相同，则添加关系数据sametable
                        if (next_file_name == current_file_name):
                            sametable_values.append((current_evidence_name, next_evidence_name))
                        # 如果遍历到的行的storage folder与当前行相同，则添加关系数据samedatabase
                        if (next_folder_name == current_folder_name):
                            samedatabase_values.append((current_evidence_name, next_evidence_name))
                    result_writer.writerow({
                        'node_evidence': current_row['node_evidence'],
                        'evidence_type': current_row['evidence_type'],
                        'edge_relationship_sameline': current_row['edge_relationship_sameline'],
                        'edge_relationship_sametable': sametable_values,
                        'edge_relationship_samedatabase': samedatabase_values

                    })
                


                
                      
```

```python
step2_output_file = "/home/hzhou/llama2/llama/experiment_result/graphstep2.csv"
transfer_graph_step2_sametable_samedatabase(step1_output_file, step2_output_file)
```
